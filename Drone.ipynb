```python
!pip install ultralytics

import os
import cv2
from ultralytics import YOLO

# ==========================
# 1) تحميل المودل
# ==========================
model = YOLO("yolov8x.pt")   # أعلى دقة

# ==========================
# 2) مسار الفيديو الذي رفعته إلى كولاب
# ==========================
input_video = "/content/mixkit-two-avenues-with-many-cars-traveling-at-night-34562-full-hd.mp4"

# ==========================
# 3) مسار حفظ الفيديو الناتج
# ==========================
output_video = "/content/processed_mixkit-two-avenues-with-many-cars-traveling-at-night-34562-full-hd.mp4"

# ==========================
# 4) الفئات المطلوبة
# ==========================
CAR_CLASSES = {2, 3, 5, 7}   # car, motorcycle, bus, truck
PERSON_CLASS = {0}           # person

# ==========================
# 5) نسبة الزحمة
# ==========================
def crowd_level(num_people, num_cars):
    total = num_people + num_cars
    if total == 0:
        return 0
    return min(100, (total / 20) * 100)  # تعديل الرقم حسب زحمة الفيديو


# ==========================
# 6) معالجة الفيديو
# ==========================
cap = cv2.VideoCapture(input_video)

fps = int(cap.get(cv2.CAP_PROP_FPS))
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer = cv2.VideoWriter(
    output_video,
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (w, h)
)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    result = model(frame, verbose=False)[0]

    num_people = 0
    num_cars = 0

    for box in result.boxes:
        cls = int(box.cls[0])
        if cls in PERSON_CLASS:
            num_people += 1
        elif cls in CAR_CLASSES:
            num_cars += 1

    congestion = crowd_level(num_people, num_cars)

    annotated = result.plot()

    cv2.putText(annotated, f"People: {num_people}",
                (20, 50), cv2.FONT_HERSHEY_SIMPLEX,
                1.1, (0, 255, 0), 3)

    cv2.putText(annotated, f"Vehicles: {num_cars}",
                (20, 100), cv2.FONT_HERSHEY_SIMPLEX,
                1.1, (255, 255, 0), 3)

    cv2.putText(annotated, f"Congestion: {congestion:.1f}%",
                (20, 150), cv2.FONT_HERSHEY_SIMPLEX,
                1.1, (0, 165, 255), 3)

    writer.write(annotated)

cap.release()
writer.release()

print("✔️ تم إنتاج الفيديو — المسار:")
print(output_video)
```
